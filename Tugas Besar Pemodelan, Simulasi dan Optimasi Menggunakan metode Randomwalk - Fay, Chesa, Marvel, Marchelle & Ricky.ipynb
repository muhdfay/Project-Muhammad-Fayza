```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Fungsi untuk membuat simulasi random walk
def random_walk_simulation(initial_price, days, volatility):
    prices = [initial_price]
    for day in range(1, days):
        price = prices[-1] * (1 + np.random.normal(0, volatility))
        prices.append(price)
    return prices

# Asumsi data historis sektor kesehatan untuk tahun 2023
initial_price_healthcare = 100  # Harga awal saham sektor kesehatan
days_healthcare = 252  # Jumlah hari perdagangan dalam satu tahun
volatility_healthcare = 0.02  # Volatilitas harian

# Simulasi pergerakan harga saham sektor kesehatan
healthcare_prices = random_walk_simulation(initial_price_healthcare, days_healthcare, volatility_healthcare)

# Visualisasi hasil simulasi
plt.plot(healthcare_prices)
plt.title('Simulasi Pergerakan Harga Saham Sektor Kesehatan')
plt.xlabel('Hari')
plt.ylabel('Harga Saham')
plt.show()
```

```python
# Asumsi data historis sektor properti untuk tahun 2023
initial_price_property = 150  # Harga awal saham sektor properti
days_property = 252  # Jumlah hari perdagangan dalam satu tahun
volatility_property = 0.03  # Volatilitas harian

# Simulasi pergerakan harga saham sektor properti
property_prices = random_walk_simulation(initial_price_property, days_property, volatility_property)

# Visualisasi hasil simulasi
plt.plot(property_prices)
plt.title('Simulasi Pergerakan Harga Saham Sektor Properti')
plt.xlabel('Hari')
plt.ylabel('Harga Saham')
plt.show()
```
